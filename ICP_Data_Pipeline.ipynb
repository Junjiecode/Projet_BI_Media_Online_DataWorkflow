```python
# define function : fetch criteo ads data

def fetch_criteo_data(ACCESS_TOKEN, AD_ACCOUNT_ID, start_date, end_date, metrics):
    url = "https://api.criteo.com/2025-01/statistics/report"

    headers = {
        'Authorization': f'Bearer {ACCESS_TOKEN}',
        'Content-Type': 'application/json'
    }

    payload = {
        "advertiserIds": AD_ACCOUNT_ID,
        "startDate": start_date,
        "endDate": end_date,
        "dimensions": ["Day", "Campaign", "Adset", "Ad"],
        "metrics": metrics,
        "format": "json",  # Specifying the format
        "currency": "EUR",  # Specifying the currency
        "timezone": "UTC"
    }
    
    # Make the request
    response = requests.post(url, headers=headers, json=payload)
    response_data = response.json()

    return response_data
```

```python
# 2) fetch_campaign_metadata : fetch all campaign data (campaign_id & campaign_name) under this ad account (org_id)

def fetch_campaign_metadata_apple(access_token, org_id):

  org_id = 1139200
  import requests
  import pandas as pd

  # === Endpoint ===
  url = "https://api.searchads.apple.com/api/v5/campaigns"

  # === Headers ===
  headers = {
      "Authorization": f"Bearer {access_token}",
      "X-AP-Context": f"orgId={org_id}",
      "Content-Type": "application/json"
  }

  # === Make the API call ===
  response = requests.get(url, headers=headers)

  # === Handle the response ===
  if response.status_code == 200:
      campaigns = response.json().get("data", [])
      df = pd.DataFrame(campaigns)
      df.rename(columns = {"id" : "campaignId", "name" : "campaign_name"}, inplace = True)
      df = df[["campaignId", "campaign_name", "status", "servingStatus"]]
      return df
  else:
      print(f"❌ Request failed with status code {response.status_code}")
      print(response.text)
```

```python
# 4) fetch ad level reporting data  :  Fetch all ad performance data within 1 campaign

def fetch_ad_data_apple(access_token, org_id, startTime, endTime, campaign_id):
  #import packages
  import requests
  import json
  import pandas as pd

  # === Required credentials ===
  org_id = 1139200
  url = f"https://api.searchads.apple.com/api/v5/reports/campaigns/{campaign_id}/ads"
  headers = {
      "Authorization": f"Bearer {access_token}",
      "X-AP-Context": f"orgId={org_id}",
      "Content-Type": "application/json"
  }

  # Define the request payload
  payload = {
      "startTime": startTime,
      "endTime": endTime,
      "timeZone": "UTC",
      "granularity": "DAILY",
      "selector": {
          "orderBy": [
              {
                  "field": "impressions",
                  "sortOrder": "DESCENDING"
              }
          ],
          "pagination": {
              "offset": 0,
              "limit": 50
          }
      },
      "returnRecordsWithNoMetrics": False,
      "returnRowTotals": True,
      "returnGrandTotals": True
  }

  # Make the POST request
  response = requests.post(url, headers=headers, json=payload)

  # Handle the response
  if response.status_code == 200:
      data = response.json()

      # Extract the list under 'row'
      rows = data['data']['reportingDataResponse']['row']

      # Flatten each row
      flattened = []
      for row in rows:
          campaign_info = row.get('metadata', {})
          for daily in row.get('granularity', []):
              flattened.append({
                  'campaignId': campaign_info.get('campaignId'),
                  'adGroupId': campaign_info.get('adGroupId'),
                  'adName': campaign_info.get('adName'),
                  'date': daily.get('date'),
                  'impressions': daily.get('impressions'),
                  'taps': daily.get('taps'),
                  #'ttr': daily.get('ttr'),
                  #'avgCPT': float(daily.get('avgCPT', {}).get('amount', 0)),
                  #'avgCPM': float(daily.get('avgCPM', {}).get('amount', 0)),
                  'localSpend': float(daily.get('localSpend', {}).get('amount', 0)),
                  #'currency': daily.get('localSpend', {}).get('currency'),
                  #'totalInstalls': daily.get('totalInstalls'),
                  'tapInstalls': daily.get('tapInstalls')
                  #'tapInstallRate': daily.get('tapInstallRate'),
                  #'tapInstallCPI': float(daily.get('tapInstallCPI', {}).get('amount', 0)),
              })

      # Create DataFrame
      df = pd.DataFrame(flattened)
      return df

  else:
      print(f"❌ Failed to retrieve campaign performance data. Status code: {response.status_code}")
      print(response.text)
```

```python
# 1) Define functions

import re

# ─── 1. Category Keyword Rules ───────────────────────────────────────────────
# Define sub-category keywords and map each to a main category.
subcategory_keywords = {
    # sous-catégorie de Electronique
    'Accessoires informatique':      ['accessoires informatique', 'accessoire informatiqu', 'pc', 'ordinateur', 'computer', 'access-info'],
    'Consoles':                      ['console'],
    'Ordinateurs':                   ['ordinateur', 'ordi', 'pav'],
    'Photo, audio & vidéo':          ['photo', 'audio', 'video', 'camera', 'multimedia', 'photoaudiovideos', 'television', 'enceinte', 'appareil photo', 'casque', 'projecteur', 'videoprojecteur', 'ecouteur', 'accessoire'],
    'Tablettes & liseuses':          ['tablette', 'liseuse', 'liseuses', 'tablettesliseuses', 'tablettes&liseuse', 'Tablettes&Lieuse_ads_Acquisition'],
    'Téléphones & objets connectés': ['apple', 'samsung', 'huawei', 'sony', 'one plus', 'google','telephon', 'mobile', 'smartphone', "telobjcon",'objet connect', 'access-tel', 'avtge-shopping', 'pmax-shopping-only', 'range', 'fr_lbc_bdc_com_gg_part_prospecting-UAC-FeedBased', 'Leboncoin RTG - Oct2022', 'android-app-purchase', 'telephony','Acquisition Apple only', 'electronique', 'montre-connect'],
    'Electroménager' :              ['electromenager', 'aspirateur', 'congelateur', 'four', 'lave-ligne', 'lave-vaisselle', 'micro-onde', 'refrigerateur'],
    'Accessoires téléphone & Objets connectés' : ['objets connect'],
    'Jeux vidéo' :                   ['jeuxvideo'],
    'Services de réparation électroniques' : ["service", "reparation"],

    # sous-catégorie de loisirs
    'Antiquités':                    ['antiquite'],
    'Artistes & Musiciens' :         ["artiste", "musicien"],
    'Billetterie' :                  ['billetterie'],
    'Collection' :                   ['collection'],
    'CD / Musique' :                 ['CD'],
    'DVD / Films'  :                 ['DVD'],
    'Instruments de musique':        ['instrument musique', 'instrument', 'musique'],
    'Livres':                        ['livre'],
    'Modélisme':                     ['modelisme', 'fr_lbc_all_com_ggshop_part_prospecting-pmax-shopping-modelisme__pr_hmq____'],
    'Vins & Gastronomie':            ['vin', 'gastronomie'],
    'Jeux & Jouets':                 ['jeu', 'jouet', 'jeuxjouets', 'jeux de societe', 'poupees', 'porteur', 'trotteur', 'draisienne', 'cusine', 'dinettes', 'jeux de construction', 'voitures et circuit', 'puzzle'],
    'Sport & plein air':             ['sport', 'plein air', 'loisir', 'sportoutdoor', 'sportpleinair'],
    'Vélos':                         ['velo', 'velos', 'cycling', 'vtt', 'vtc'],
    'Équipements vélos':             ['equipement velo', 'equip-velo'],

    # sous-catégorie de Maison & Jardin
    'Ameublement':                   ['ameublement', 'armoire', 'buffet', 'canape', 'chaise', 'tabouret', 'banc', 'fauteuil', 'meuble', 'table'],
    'Bricolage':                     ['bricolage'],
    'Décoration':                    ['decoration', 'applique', 'horloge', 'pendule', 'lampadaire', 'lampe', 'lustre', 'miroir', 'rideau', 'voilage', 'store', 'sculpture', 'statue', 'suspension', 'tableau', 'toile', 'tapis', 'vase', 'cache pot', 'ceramique'],
    'Jardin & Plantes':              ['jardin', 'plante', 'jardinplantes'],
    'Linge de maison' :              ['lingemaison', 'textile', 'linge'],
    'Services de jardinerie & bricolage' : ["service de jardin"],
    'Arts de la table' :             ['artsdelatable', 'assiette', 'service de vaisselle', 'verre'],
    'Papeterie & Fournitures scolaire' : ["papeterie", "fourniture scolaire"],

    # sous-catégorie de Famille
    'Equipement bébé' :              ['equip-bebe', 'poussette', 'siege auto'],
    'Mobilier enfant'  :             ['mobilierenfant', 'baignoire', 'chaise haute', 'lit bebe'],
    'Vêtements bébé' :               ['vetementsbebe'],

    # sous-catégorie de Mode
    'Vêtements':                     ['vetement', 'mode'],
    'Chaussures' :                   ['chaussures'],
    'Montres & Bijoux':              ['montre', 'bijou', 'montresbijoux'],
    'Accessoires & Bagagerie' :      ['access-bagagerie'],


    # sous-catégorie : Autres
    'Animaux' :                      ['animaux'],
    'Accessoires animaux' :          ['accessoiresanimaux']
}

category_map = {
                "Electronique" : ['Accessoires informatique', 'Consoles', 'Ordinateurs', 'Photo, audio & vidéo', 'Tablettes & liseuses', 'Téléphones & objets connectés', 'Électroménager', 'Accessoires téléphone & Objets connectés', 'Jeux vidéo', 'Services de réparation électroniques'],
                "Loisir" : ['Antiquités', 'Artistes & Musiciens', 'Billetterie', 'Collection', 'CD / Musique', 'DVD / Films', 'Instruments de musique', 'Livres', 'Modélisme', 'Vins & Gastronomie', 'Jeux & Jouets', 'Sport & plein air', 'Vélos', 'Équipements vélos'],
                "Maison & Jardin" : ["Ameublement", "Bricolage", "Décoration", "Jardin & Plantes", "Linge de maison", "Services de jardinerie & bricolage", "Arts de la table", "Papeterie & Fournitures scolaire"],
                "Famille" : ["Equipement bébé", "Mobilier enfant", "Vêtements bébé"],
                "Mode" : ["Vêtements", "Chaussures", "Montres & Bijoux", "Accessoires & Bagagerie"],
                "Autres" : ["Animaux", "Accessoires animaux"]
                }


# ─── 2. Helper Function: Normalize Text ────────────────────────────────────────
import unicodedata
def normalize(text: str) -> str:
    """
    Remove accents and convert text to lowercase.
    Used to prepare strings for keyword-based categorization.
    """
    if pd.isna(text):
        return ''
    # Decompose unicode into base characters + combining marks
    nfkd = unicodedata.normalize('NFD', text)
    # Discard combining marks and lowercase
    return ''.join(c for c in nfkd if not unicodedata.combining(c)).lower()

# ─── 3. Assign category & sub_category ────────────────────────────────────────

# Assign sub-category based on tokens
def assign_sub_category(text: str) -> str:
    """
    Match normalized text against subcategory keywords using regex word boundaries.
    """
    for sub_category, keywords in subcategory_keywords.items():
        for kw in keywords:
            if re.search(kw, text):
                return sub_category
    return 'Uncategorized'

# Assign Category based on Sub-category
def assign_category(sub_category):
    for category, sub_categories in category_map.items():
        if sub_category in sub_categories:
            return category
    return 'Uncategorized'
```
